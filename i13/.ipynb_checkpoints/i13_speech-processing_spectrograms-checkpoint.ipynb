```python
TRAIN_WAVFILES_PATH = './train/audio/'
TRAIN_SPECT_PATH = './train/spect/'
TRAIN_SPECT_TMP_PATH = './train/spect_tmp_2/'
PREDICTION_LABELS = ["yes", "no", "up", "down", "left", "right", "on", "off", "stop", "go", "unknown"]
```

```python
subfolder_names_list = [x for x in os.listdir(train_audio_folder) if os.path.isdir(train_audio_folder + '/' + x)]
```

```python
def create_label_subfolders(path, subfolder_names):
    for x in subfolder_names:
        subfolder_path = path + x
        if os.path.exists(subfolder_path):
            print("%s already exists in this folder" % x)
        else: 
            os.makedirs(subfolder_path)
            print("%s folder created at %s" % (x, subfolder_path))
```

```python
create_label_subfolders(TRAIN_SPECT_TMP_PATH, PREDICTION_LABELS)
```

```python
def log_specgram(audio, sample_rate, window_size=20, step_size=10, eps=1e-10):
    nperseg = int(round(window_size * sample_rate / 1e3))
    noverlap = int(round(step_size * sample_rate / 1e3))
    freqs, _, spec = signal.spectrogram(audio, fs=sample_rate, window='hann', 
                                        nperseg=nperseg, noverlap=noverlap,
                                    detrend=False)
    return freqs, np.log(spec.T.astype(np.float32) + eps)
```

```python
def create_spectrogram_file(subfolder, filename):
    tmp_img_path = TRAIN_SPECT_TMP_PATH + subfolder + '/' + filename[:-4] + "_tmp.jpg"
    spect_img_path = TRAIN_SPECT_PATH + subfolder + '/' + filename[:-4] + "_spect.jpg"
    if subfolder not in PREDICTION_LABELS:
        tmp_img_path = TRAIN_SPECT_TMP_PATH + 'unknown/' + filename[:-4] + "_tmp.jpg"
        spect_img_path = TRAIN_SPECT_PATH + 'unknown/' + filename[:-4] + "_spect.jpg"
    sample_rate, samples = wavfile.read(TRAIN_WAVFILES_PATH + subfolder + '/' + filename)
    _, spectrogram = log_specgram(samples, sample_rate)
    plt.imsave(tmp_img_path, spectrogram.T, origin='lower')
    spect_img = Image.open(tmp_img_path).resize([96,96])
    spect_img.save(spect_img_path)
```

```python
%%time
total = 0
for x in subfolder_names_list:
    subfolder_wav_files = [x for x in os.listdir(train_audio_folder + x) if '.wav' in x]
    len_subfolder = len(subfolder_wav_files)
    total += len_subfolder
    for y in range(len_subfolder):
        create_spectrogram_file(x, subfolder_wav_files[y])
        if x not in PREDICTION_LABELS and y > 200:
            break;
        if y % 100 == 0:
            print("%d of %d images in %s processed" % (y, len_subfolder, x))
    print('file count in %s: %d' % (x, len_subfolder))
print('total file count: %d' % (total))
```
